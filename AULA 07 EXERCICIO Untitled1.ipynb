# Calculadora Modular em Python

```python
def adicionar(num1: float | int, num2: float | int) -> float | int:
    """Realiza a soma de dois números."""
    return num1 + num2

def subtrair(num1: float | int, num2: float | int) -> float | int:
    """Realiza a subtração de dois números."""
    return num1 - num2

def multiplicar(num1: float | int, num2: float | int) -> float | int:
    """Realiza a multiplicação de dois números."""
    return num1 * num2

def dividir(num1: float | int, num2: float | int) -> float | str:
    """Realiza a divisão de dois números, verificando divisão por zero."""
    if num2 == 0:
        return "Erro: Não é possível dividir por zero!"
    return num1 / num2
```

## Exemplos de Uso

```python
# Exemplos com inteiros
print(f"10 + 5 = {adicionar(10, 5)}")
print(f"10 - 5 = {subtrair(10, 5)}")
print(f"10 * 5 = {multiplicar(10, 5)}")
print(f"10 / 5 = {dividir(10, 5)}")

# Exemplos com floats
print(f"10.5 + 2.3 = {adicionar(10.5, 2.3)}")
print(f"10.5 - 2.3 = {subtrair(10.5, 2.3)}")
print(f"10.5 * 2.3 = {multiplicar(10.5, 2.3)}")
print(f"10.5 / 2.5 = {dividir(10.5, 2.5)}")

# Exemplo de divisão por zero
print(f"10 / 0 = {dividir(10, 0)}")
```

## Calculadora Interativa (Clean Code)

Esta seção implementa a calculadora com um loop interativo, tratamento de entrada para diferentes separadores decimais (vírgula e ponto) e uma opção para sair do programa.

```python
def get_numeric_input(prompt_message: str) -> float | str:
    """Obtém uma entrada numérica do usuário, tratando vírgulas/pontos decimais e a opção 'sair'."""
    while True:
        user_input = input(prompt_message).strip().lower()
        if user_input == 'sair':
            return 'sair'
        try:
            # Substitui vírgula por ponto para permitir a conversão para float
            cleaned_input = user_input.replace(',', '.')
            return float(cleaned_input)
        except ValueError:
            print("Entrada inválida. Por favor, digite um número válido ou 'sair'.")

def get_operation_input() -> str:
    """Obtém a operação desejada do usuário ou a opção 'sair'."""
    while True:
        op = input("Escolha uma operação (+, -, *, /) ou digite 'sair': ").strip().lower()
        if op in ['+', '-', '*', '/', 'sair']:
            return op
        else:
            print("Operação inválida. Por favor, escolha +, -, *, / ou 'sair'.")

def calculadora_interativa():
    """Função principal da calculadora interativa."""
    print("\n--- Bem-vindo à Calculadora Modular! ---")
    print("Digite 'sair' a qualquer momento para encerrar.")

    while True:
        num1 = get_numeric_input("Digite o primeiro número: ")
        if num1 == 'sair':
            break

        operacao = get_operation_input()
        if operacao == 'sair':
            break

        num2 = get_numeric_input("Digite o segundo número: ")
        if num2 == 'sair':
            break

        resultado = None
        if operacao == '+':
            resultado = adicionar(num1, num2)
        elif operacao == '-':
            resultado = subtrair(num1, num2)
        elif operacao == '*':
            resultado = multiplicar(num1, num2)
        elif operacao == '/':
            resultado = dividir(num1, num2)
        else:
            # Esta linha não deve ser alcançada se get_operation_input funcionar corretamente
            print("Erro interno: operação desconhecida.")
            continue

        print(f"Resultado: {num1} {operacao} {num2} = {resultado}\n")

    print("--- Calculadora encerrada. Até mais! ---")
```

```python
# Inicia a calculadora interativa
calculadora_interativa()
```
